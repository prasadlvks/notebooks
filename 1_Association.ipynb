```python
francedf = df[df.Country=='France']
```

```python
francedf
```

```python
unstacked=francedf.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack()
```

```python
filleddf=unstacked.reset_index().fillna(0)
```

```python
filleddf
```

```python
basket=filleddf.set_index('InvoiceNo')
```

```python
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

```python
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```
