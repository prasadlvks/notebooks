ETS framework - Error, Trend, and Season

Additive: A

Multiplicative: M

None: N (Not included)

ETS(M,N,N): Simple exponential smoothing with multiplicative errors.

ETS(A,A,N): Holt’s linear method with additive errors.

ETS(M,A,N): Holt’s linear method with multiplicative errors.

In general, additive terms show linear or constant behavior.

if the trend shows an exponential variation, either positive or negative, then it has a multiplicative behavior.

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from matplotlib  import pyplot
```

```python
index = pd.date_range('11/07/2019', periods=35, freq='D')
df = pd.read_csv('Murali.csv',parse_dates=True)
```

```python
df.head()
```

```python
df.set_index(index, inplace=True)
```

```python
df.head()
```

```python
df.drop('Date', axis=1, inplace=True)
```

```python
df.tail()
```

```python
pyplot.plot(df.index,df.DBSize)
```

```python
decomposition = seasonal_decompose(df, model='additive')
decomposition.plot()
```

```python
model = SimpleExpSmoothing(df.values)
```

```python
fit1 = model.fit(smoothing_level=.2)
```

```python
pred1 = fit1.forecast(2)
```
