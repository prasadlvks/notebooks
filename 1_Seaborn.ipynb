# Seaborn 
## distplot
## jointplot
## pairplot - will plot pairwise relationships accross an entire dataframe
## rugplot
## kdeplot

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],color='red')
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],bins=30)
```

```python
sns.jointplot(tips['total_bill'],tips['size'])
```

```python
sns.jointplot(tips['total_bill'],tips['size'],kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.countplot(x='size', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips)
```

```python
sns.factorplot(x='sex',y='total_bill', data=tips)
```

```python
sns.factorplot(x='sex',y='total_bill', data=tips, kind='violin')
```

```python
tipscorr = tips.corr()
```

```python
tipscorr
```

```python
sns.heatmap(tipscorr)
```

```python
tips.pivot_table(values='tip',columns='sex')
```

```python
tips.pivot_table(values='total_bill',columns='sex')
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips)
```
