# Gain at 95% confidence interval - 95% CI

```python
from scipy import stats
glaxo_df_ci = stats.norm.interval(.95, loc=glaxo_df.gain.mean(), scale=glaxo_df.gain.std())
glaxo_df_ci
```

```python
beml_df_ci = stats.norm.interval(.95, loc=beml_df.gain.mean(), scale=beml_df.gain.std())
beml_df_ci
```

# Probability of making 2% loss - Cumulative distribution function - CPD

```python
stats.norm.cdf(-.02, loc=glaxo_df.gain.mean(), scale=glaxo_df.gain.std())
```

```python
stats.norm.cdf(-.02, loc=beml_df.gain.mean(), scale=beml_df.gain.std())
```

# Probability of making 2% gain

```python
1 - stats.norm.cdf(.02, loc=glaxo_df.gain.mean(), scale=glaxo_df.gain.std())
```

```python
1 - stats.norm.cdf(.02, loc=beml_df.gain.mean(), scale=beml_df.gain.std())
```
