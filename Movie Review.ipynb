```python
data[(data.duration >= 200) & (data.genre=='Drama')]
```

```python
data[(data.duration <= 200) & (data.genre.isin(['Crime','Drama','Action']))].head()
```

```python
#using Agg function
data.groupby(['genre']).star_rating.agg(['count','min','max','mean','std']).sort_values('count',ascending=False).head()
```

```python
data.groupby('title').title.count().sort_values(ascending=False).head()
```

```python
#using Agg function
data.groupby(['content_rating','genre']).star_rating.agg(['count','min','max','mean','std']).sort_values('count',ascending=False).head()
```

```python
data.groupby('content_rating').title.count()
```

```python
data[data['star_rating']==data['star_rating'].min()]
```

```python
data.groupby('star_rating').star_rating.count().sort_values(ascending=False)
```

```python
data.head()
```

```python
#what is Highest start_ratings by Content_rating.
print('Max',data.groupby('content_rating').star_rating.max(),'\n')
print('Min',data.groupby('content_rating').star_rating.min(),'\n')
print('Mean',data.groupby('content_rating').star_rating.mean(),'\n')
print('Count',data.groupby('content_rating').star_rating.count(),'\n')
print('Std',data.groupby('content_rating').star_rating.std(),'\n')
```

```python
#what is Highest start_ratings by genre.
print('Max',data.groupby('genre').star_rating.max().head(),'\n')
#Min, Mean, Count, Std
```

```python
#what is Highest start_ratings by content_rating and genre.
print('Max',data.groupby(['content_rating','genre']).star_rating.max(),'\n')
```
